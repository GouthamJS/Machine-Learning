```python
x=df.loc[:,['sepal.length', 'sepal.width', 'petal.length', 'petal.width']]
x
```

```python
df['variety'].value_counts()
```

```python
y=df[ 'variety']
y
```

```python
#train test split model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.6,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train,x_test,y_train,y_test
```

```python
#EDA2 (model building and model evaluation)
#GNB Calsulates the mean and varience of each class during training phase
#During pprediction, it calculates probability of each class and assigns and instant
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_train)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_train,y_pred))
#gives diffrence between actual and predicted values
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,y_pred))
```
